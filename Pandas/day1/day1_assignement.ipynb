```python
import pandas as pd

data = {
    "Name": ["Puneet", "Rahul", "Aman", "Riya"],
    "Age": [19, 20, 21, 18],
    "Marks": [85, 72, 90, 65],
    "City": ["Jaipur", "Delhi", "Kota", "Mumbai"]
}

df = pd.DataFrame(data,index=['r1','r2','r3','r4'])

res = df[df["Marks"] > 80]

print(res)
res = df[(df["Age"] > 18) & (df["Marks"] > 70)]

print(res)

print(df.iloc[0,2])


```

```python
data = {
    "Name": ["Puneet", "Rahul", "Aman", "Riya"],
    "Age": [19, 20, 21, 18],
    "Marks": [85, 72, 90, 65],
    "City": ["Jaipur", "Delhi", "Kota", "Mumbai"]
}

df = pd.DataFrame(data,index=['r1','r2','r3','r4'])

print(df.iloc[0:3, 0:4])

df = df[df["Marks"] <= 70]

print(df)
```
