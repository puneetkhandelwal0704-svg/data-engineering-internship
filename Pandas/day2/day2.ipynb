```python
import pandas as pd
df1=pd.DataFrame({'A':[1,2,3,4],'B':[10,20,30,40]})
df2=pd.DataFrame({'A':[11,22,35],'B':[100,208,300]},index=[1,2,3])
print(df1)
print(df2)
print(df1+df2)
print(df2-df1)
print(df1['A'][1]+df2['A'][3])
```

```python
one=pd.DataFrame({
   'Name': ['A1', 'A2', 'A3', 'A4'],
   'subject':['sub1','sub2','sub4','sub6'],
   'Marks':[98,90,87,69]},
   index=[1,2,3,4])
print(one)
two = pd.DataFrame({
    'Name': ['B1', 'B2', 'B3', 'B4', 'B5'],
    'subject': ['sub2', 'sub4', 'sub3', 'sub6', 'sub5'],
    'Marks': [89, 80, 79, 97, 88]},
    index=[1, 2, 3, 4, 5])
print(two)
print("concat of two dataframes:", pd.concat([one, two],keys=['one','two']))
print("concat of two dataframes:", pd.concat([one, two],ignore_index=True))

```

```python
import pandas as pd
one=pd.DataFrame({
    'Id':[1,2,3,4],
    'Name': ['A1', 'A2', 'A3', 'A4'],
    'subject':['sub1','sub2','sub4','sub6']})
two = pd.DataFrame({
    'Name': ['B1', 'B2', 'B3', 'B4'],
    'subject': ['sub1', 'sub2', 'sub3', 'sub4'],
    'Id':[1,2,3,4]})
print(one)
print(two)
res = one.merge(two, on='Id')
print(res)
```

```python
data = {
'Course': ['CS', 'Civil', 'CS', 'Civil', 'CS'],
'Year': [1, 2, 1, 1, 2],
'Student': [100, 150, 120, 90, 180]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

pivoted = df.pivot_table(index='Course', columns="Year", values='Student', aggfunc='sum')
print(pivoted)
```
