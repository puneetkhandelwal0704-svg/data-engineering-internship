```python
import pandas as pd

df= pd.DataFrame([[9, -3, -2], [-5, 1, 8], [6, 4, -8]],
index=['a', 'c', 'd'], columns=['one', 'two', 'three'])
df=df.reindex(['a', 'b', 'c', 'd', 'e'])
print(df)
print("Forward fill")
res=df.ffill()
print(res)
f=df.fillna('-')
print(f)
```

```python
import pandas as pd

df= pd.DataFrame([[9, -3, -2], [-5, 1, 8], [6, 4, -8]],
index=['a', 'c', 'd'], columns=['one', 'two', 'three'])
df=df.reindex(['a', 'b', 'c', 'd', 'e','f'])
print(df)
res=df.ffill(limit=1)
print(res)
```

```python
#groupby
import pandas as pd
df=pd.DataFrame({'one':["A","B","A","C","B","C","B"],'two':[5,40,60,200,700,71,89],'three':['X','Y','X','Z','Y','Z','Y']})
print(df)
res=df.groupby(['one','three'])['two'].sum()
print(res)
```

```python
import pandas as pd
df = pd.DataFrame({'one':["A","B","A","C","B","C","B"], 'two':[5,40,60,200,700,71,89], 'three':['X','Y','X','Z','Y','Z','Y']})
print(df)
for name, group in df.groupby('one'):
    print(name)
    print(group)
```

```python
import pandas as pd
df=pd.DataFrame({'one':["A","B","A","C","B","C","B"],'two':[5,40,60,200,700,71,89],'three':['X','Y','X','Z','Y','Z','Y']})
print(df)
res=df.sort_values(by='one',ascending=False)
print(res)
```

```python
import pandas as pd
df=pd.DataFrame({'one':["A","B","A","C","B","C","B"],'two':[5,40,60,200,700,71,89],'three':['X','Y','X','Z','Y','Z','Y']})
print(df)
res=df.sort_values(by=['one','two'],ascending=[True,True])
print(res)
```
