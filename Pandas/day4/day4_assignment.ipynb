```python
import numpy as np
a=np.array([
    [1,2,3,4,5],
    [np.nan,22,35,43,51]
    ])
print(a,type(a))
```

```python
arr=np.nan_to_num(a,nan=0) 
print(arr)
```

```python
a_transpose=arr.T
print(a_transpose) 
```

```python
import numpy as np

arr = np.arange(24).reshape(2, 3, 4)

print("Original Shape:", arr.shape)

# move axes
new_arr = np.moveaxis(arr, 0, 2)

print("\nNew Shape:", new_arr.shape)

print("\nNew Array:\n", new_arr)
```

```python
import numpy as np
arr=np.array([
    [1,2,np.nan],
    [4,np.nan,45],
    [7,8,9]
])
print(arr)
col_mean=np.nanmean(arr,axis=0)
print("average of colums",col_mean)
ind=np.where(np.isnan(arr))
print(ind)
arr[ind] = np.take(col_mean, ind[1])
print(arr)
```
