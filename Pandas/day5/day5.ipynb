```python
import numpy as np
a=np.empty((2,4))
print(a)
```

```python
a=np.array([1,2,3,4,5])
print(a[2]+a[4])
```

```python
a=np.array([[1,2,3,4],[6,7,8,9]])
print(a)
print(a[1,3])
```

```python
a=np.array([[[1,2,3,4],[6,7,8,9]],[[11,12,13,14],[21,22,23,24]]])
print(a[1,1,-2])
```

```python
a=np.array([2,3,4,5,6,7,8])
print(a[-2:-1])  #start index is included and end index is excluded 
```

```python
a=np.array([[2,9,4,5,6,7,8],[12,13,14,15,16,17,18]])
print(a[0:2,1:4])
print("shape:",a.shape)
```

```python
a=np.zeros((3,4,5))
print(a)
```

```python
a=np.full((2,2,3), 7)
print(a)
```
