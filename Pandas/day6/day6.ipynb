```python
import numpy as np
a=np.array([[1,2,3],[7,4,9]])
print(a)
res=np.transpose(a)
print(res)
```

```python
arr1=np.array([[1,2,3],[4,5,6]])
arr2=np.array([[7,8,9],[10,11,12]])
res=np.concatenate((arr1,arr2),axis=0) 
print(res)
```

```python
arr=np.arange(9)
print(arr)
res=np.split(arr,3)
print(res)
```

```python
arr=np.array([[1,2,3],[7,4,9]])
print(arr)
b=np.resize(arr,(3,2)) 
print(b)

```

```python
import numpy as np
arr=np.array([[1,2,3],[4,5,6]])
print(np.append(arr,[[7,8,9]],axis=0))
print(np.append(arr,[[7,8,9],[10,11,12]],axis=1))
```

```python
arr=np.array([[1,2],[3,4],[7,8],[9,10]])
print(arr)
a=np.insert(arr,2,[5,6],axis=0)
print(a)
```

```python
arr=np.array([1,2,3,4,5])
res=np.flip(arr) 
print(res)
```

```python
arr=np.array([5,3,4,2,8,7,1])
arr=np.sort(arr)
print(arr)
```

```python
arr=np.array([5,3,4,2,8,7,1])
arr=np.argmin(arr)
print(arr)
```

```python
#indexing and slicing

arr=np.arange(10)
b=arr[6]
print(b)
```

```python
a=np.array(
    [
        [[123,4,5],[2,3,4]],
        [[123,4,5,],[2,3,4]]
    ]
) 
print(a)
```

```python
import numpy as np

a=np.array([[1,2,3],[4,5,7]])
b=np.array([[7,8,9],[10,11,12]])
print(a+b) 
print(a-b)
print(a*b)
print(a/b)
print(a**b)
```

```python
b=np.array([[7,8,9],[10,11,12]])
res=np.sum(b,axis=1)
print(b)
print(res)
```

```python
a=np.array([[7,8,9],[10,11,12]])
res=np.mean(a,axis=0)
print(res)  
```
