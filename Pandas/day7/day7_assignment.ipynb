```python
arr=np.array([
    [6,-8,73,-110],
    [np.nan,-8,0,94]
])
print("orignal array:",arr)

res=np.nanmean(arr,axis=0)
inds=np.where(np.isnan(arr))
arr[inds]=np.take(res,inds[1])
print(arr)
```

```python
import numpy as np
from statistics import mode

arr1 = np.array([
    [10, 20],
    [30, 40]
])

arr2 = np.array([
    [50, 60],
    [70, 80]
])

# Average Array
avg = (arr1 + arr2) / 2

print("Average Array:")
print(avg)

combined = np.concatenate((arr1.flatten(), arr2.flatten()))

print("\nMean:", np.mean(combined))
print("Median:", np.median(combined))
print("Mode:", mode(combined))
```

```python
arr1=np.array([[1,-2,3],
               [-1,3,-1],
               [2,-5,5]])

arr2=np.array([9,-6,18])
solution=np.linalg.solve(arr1,arr2)
print(solution)
print("x =", solution[0])
print("y =", solution[1])
print("z =", solution[2])
```

```python
import matplotlib.pyplot as plt
import numpy as np

subjects = ['Maths', 'Python', 'DBMS', 'OS', 'Java']

sem1 = [75, 82, 78, 70, 85]
sem2 = [82, 88, 85, 80, 90]

x = np.arange(len(subjects))

plt.figure(figsize=(8,5))

plt.plot(x, sem1,
         marker='o',
         linestyle='--',
         color='blue',
         label='Semester 1')

plt.plot(x, sem2,
         marker='s',
         linestyle='-',
         color='red',
         label='Semester 2')

plt.title("Semester Result Comparison")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.xticks(x, subjects)
plt.grid(True)
plt.legend()

plt.show()
```
